```python
###Data Cleaning##
```

```python
data.isnull().sum()
```

```python
### dropping all missing values rows
data.dropna(inplace=True)
data.isnull().sum()
```

```python
## dealing with duplicated rows 
data.duplicated().sum()
data[data.duplicated()]
## deleting all duplicated
data.drop_duplicates(inplace=True)

data.duplicated().sum()
```

```python
data.dtypes
```

```python
## Changing data types
data["id"]=data['id'].astype(object)
data["host_id"]=data['host_id'].astype(object)
data.dtypes
```

```python
## Data Analysis 
## Univariate Analysis
```

```python
## Price distribution
data['price']
```

```python
# identifying outliers in price

df = data[data['price']<1500]
sns.boxplot(data=df, x='price')
```

```python
plt.figure(figsize=(8,5))
plt.title("Price Distribution")
sns.histplot(data=df,x='price',bins=100)
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
## Availability of 365
plt.figure(figsize=(8,5))
plt.title("Availability_365 Distribution")
sns.histplot(data=df,x='availability_365')
plt.ylabel('Frequency')
plt.show()
```

```python
df.dtypes
```

```python
## Price per bed
df['price_per_bed']=df['price']/df['beds']
```

```python
## average price per bed
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
## Bivariate Analysis
```

```python
## Price dependency on neighbourhood
fig, ax = plt.subplots(figsize=(10, 6))

sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type', ax=ax)

ax.grid(False)

plt.tight_layout()

plt.show()
```

```python
## Number of reviews and price relation
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group',ax=ax)
ax.grid(False)
plt.title("Locality and Review Dependency")
plt.tight_layout()

plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
#Geographical Distribution of Airbnb listing
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type', ax=ax)

ax.grid(False)
plt.title("Geographical Distribution of Airbnb Listing")
plt.tight_layout()

plt.show()
```

```python
#heatmap - correlation of one variable with others for numerical column
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']]
corr
```

```python
corr = df.corr(numeric_only=True)   # ensures only numeric columns

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", cbar=True, square=True)

plt.title("Correlation Heatmap", fontsize=16, weight="bold")
plt.show()
```
